```python
column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
df.plot(kind='box', subplots=True, layout=(2,5),figsize=(30,20),color='red')
plt.show()
```

```python
# Exclude categorical columns
numeric_df = df.drop(columns=categorical_cols)

# Compute the correlation matrix
corr = numeric_df.corr()

# Print the correlation matrix
print(corr)

# Filter correlations with absolute value >= 0.90
corr = corr[abs(corr) >= 0.90]

#Plot the heatmap
sns.heatmap(corr,annot=True,cmap='Set3',linewidths=0.2)
plt.show()
```

```python
high_corr_cols = ['emp.var.rate','euribor3m','nr.employed']
```

```python
df1 = df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1)  # axis=1 indicates columns
df1.columns
```

```python
df1.shape
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded['deposit'].value_counts()
```

```python
x = df_encoded.drop('deposit',axis=1)  # independent variable
y = df_encoded['deposit']              # dependent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
#SPLITTING ALL DATA IN TWO PARTS(TRAIN AND TEST)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#EVALUATING PERFORMANCE MEASURES

from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc = accuracy_score(y_test,y_pred)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    print('Training Score',train_score)  
    print('Testing Score',test_score)
```

```python
#FITTING TO A DECISION TREE
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
#ACCURACY ON TRAIN AND TEST DATA
mscore(dt)
```

```python
ypred_dt = dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```

```python
#PLOTTING DECISION TREE
from sklearn.tree import plot_tree
cn = ['no','yes']
fn = x_train.columns
print(fn)
print(cn)
```

```python
plt.figure(figsize=(40,15))
plot_tree(dt,class_names=cn,filled=True)
plt.show()
```

```python
dt1 = DecisionTreeClassifier(criterion='entropy',max_depth=4,min_samples_split=15)
dt1.fit(x_train,y_train)
```

```python
mscore(dt1)
```

```python
ypred_dt1 = dt1.predict(x_test)
eval_model(y_test,ypred_dt1)
```
